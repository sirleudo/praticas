# Gráfico de Dispersão (Scatter Plot) 

## Plotar o Scatter Plot

```python
# importar os pacotes necessários
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# importar o dataset
DATA_PATH = "https://raw.githubusercontent.com/carlosfab/dsnp2/master/datasets/housing.csv"
df = pd.read_csv(DATA_PATH)
```

```python
# ver as primeiras entradas
df.head()
```

```python
# plotar o scatter plot
fig, ax = plt.subplots()

# plotar median_income vs. median_house_value
df.plot.scatter(x = 'median_income', y = 'median_house_value', ax = ax, s= .5)
plt.show()
```

## Regressão Linear em um Scatter Plot

```python
# importar o seaborn
import seaborn as sns
```

```python
# plotar scatter plot e regressão linear
fig, ax = plt.subplots()
ax.set_xlim(right=8)
sns.regplot('median_income', 'median_house_value', scatter_kws={"s": 0.2},
           line_kws={"color": "red"}, data=df, ax = ax)

plt.tight_layout()
```
